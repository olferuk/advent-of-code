```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt

data = Path("data/09.txt").read_text().splitlines()
```

```python
T, H = (0, 0), (0, 0)
visited = []

directions = {
    "R": (0, 1),
    "L": (0, -1),
    "U": (-1, 0),
    "D": (1, 0),
}

for line in data:
    d, n = line.split()
    v = directions[d]
    for _ in range(int(n)):
        H = (H[0] + v[0], H[1] + v[1])

        dist = max(abs(H[0] - T[0]), abs(H[1] - T[1]))

        if dist > 1:
            dy, dx = H[0] - T[0], H[1] - T[1]
            dy, dx = np.clip(dy, -1, 1), np.clip(dx, -1, 1)
            T = (T[0] + dy, T[1] + dx)
        visited.append(T)

len(set(visited))
```

```python
plotting = np.array(visited)

plt.figure(figsize=(12, 8), dpi=100)
plt.plot(plotting[:, 0], plotting[:, 1], c="orange")
```
