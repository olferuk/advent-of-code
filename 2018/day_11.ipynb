```python
%load_ext lab_black

from puzzles import load

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
int(load(11).strip())
```

```python
A = np.zeros([300, 300], dtype=np.int32)
serial = 1723

for y in range(1, 301):
    for x in range(1, 301):
        rid = x + 10
        power = rid * y
        power = power + serial
        rid = rid * power
        hundreds = rid // 100 % 10
        A[y - 1, x - 1] = hundreds - 5

plt.figure(figsize=(12, 12), dpi=300)
sns.heatmap(A, ax=plt.gca(), square=True)
```

```python
s_max = -1
for y in range(1, 301 - 2):
    for x in range(1, 301 - 2):
        s = A[y : y + 3, x : x + 3].sum()
        if s > s_max:
            s_max = s
            c_best = (x, y)

s_max, (c_best[0] + 1, c_best[1] + 1)
```

```python
from tqdm import tqdm
```

```python
s_max = -1
for k in tqdm(range(3, 301)):
    for y in range(len(A) - k + 1):
        for x in range(len(A[0]) - k + 1):
            s = A[y : y + k, x : x + k].sum()
            if s > s_max:
                s_max = s
                c_best = (x, y, k)

s_max, (c_best[0] + 1, c_best[1] + 1, c_best[2])
```
