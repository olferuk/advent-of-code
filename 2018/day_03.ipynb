```python
import numpy as np
import re
```

```python
cmds = np.array(
    [re.findall(r"\d+", line) for line in load(3).strip().split("\n")]
).astype(int)
```

```python
A = np.zeros([1000, 1000], np.uint8)

for cmd in cmds:
    A[cmd[2] : cmd[2] + cmd[-1], cmd[1] : cmd[1] + cmd[-2]] += 1

(A > 1).sum()
```

---

```python
for cmd in cmds:
    patch = A[cmd[2] : cmd[2] + cmd[-1], cmd[1] : cmd[1] + cmd[-2]]
    if patch.sum() == cmd[-2] * cmd[-1]:
        print(cmd[0])
        break
```
